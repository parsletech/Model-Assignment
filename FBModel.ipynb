````python
for col in data1.columns:
    if data1[col].dtype == 'float64' or data1[col].dtype == 'int64':
        data1[col].fillna(data1[col].mean(), inplace=True)
    else:
        data1[col].fillna(data1[col].mode()[0], inplace=True)
#Replaces the missing values with the mean for numeric values and the mode for the string
````

````python
numeric_cols = data1.select_dtypes(include=['float64', 'int64']).columns

for col in numeric_cols:
    q1 =data1[col].quantile(0.25)
    q3 =data1[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    outliers = data1[(data1[col] < lower_bound) | (data1[col] > upper_bound)]
    if not outliers.empty:
        print(f'Outliers found in column {col}:')
        print(outliers)
        #this code checks for outliers in the numeric columns.
````

````python
for col in numeric_cols:
    q1 =data1[col].quantile(0.25)
    q3 = data1[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    data1[col] = data1[col].apply(lambda x: data1[col].median() if x < lower_bound or x > upper_bound else x)
#This code replaces outliers with median.
````

##Data preparation.

###Data Separation

````python
y = data1['comment']
y
````

````python
x = data1.drop('comment' , axis = 1)
x
````

````python
x['Type'].unique()
#this is the type column
````

````python
Types = ['Photo', 'Status', 'Link', 'Video']
````

````python
from sklearn.preprocessing import OrdinalEncoder
oenc = OrdinalEncoder(categories = [Types])

x['Type'] = oenc.fit_transform(x[['Type']])
#I have converted the type column into ordinal data
````

````python
x.head()
````

````python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=40)

````

##Model building




###Linear Regression

````python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)

y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
y_lr_test_pred
````
